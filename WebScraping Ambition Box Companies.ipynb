```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
webpage
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
company=soup.find_all('div',class_="company-content-wrapper")
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
company_type=[]
hq=[]
age=[]
employee_count=[]

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    age.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    employee_count.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    
```

```python
pd.DataFrame({'Company_Name':name,
              'Company_Type':company_type,
              'HeadQuater':hq,
              'Company_Age':age,
              'Total_employee':employee_count,
              'reviews':reviews,
              'ratings':rating
              })
```

```python
final=pd.DataFrame()
for j in range(1,1000):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_="company-content-wrapper")
    name=[]
    rating=[]
    reviews=[]
    company_type=[]
    hq=[]
    age=[]
    employee_count=[]
    
    for i in company:
        
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)
        
        try:
            rating.append(i.find('p',class_='rating').text.strip())
        except:
            rating.append(np.nan)
        
        try:
            reviews.append(i.find('a',class_='review-count').text.strip())
        except:
            reviews.append(np.nan)
        
        try:
            company_type.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            company_type.append(np.nan)
        
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        
        try:
            age.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            age.append(np.nan)
        
        try:
            employee_count.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employee_count.append(np.nan)
        
        
    df=pd.DataFrame({'Company_Name':name,
              'Company_Type':company_type,
              'HeadQuater':hq,
              'Company_Age':age,
              'Total_employee':employee_count,
              'reviews':reviews,
              'ratings':rating
              })
    final=final.append(df ,ignore_index=True)

    

    

    
```
